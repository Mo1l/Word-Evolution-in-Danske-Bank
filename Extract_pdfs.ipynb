### Extracting pdfs and converting to txt 

```python
from bs4 import BeautifulSoup
import requests
from selenium import webdriver
import re, os
from tqdm.notebook import tqdm
import time
```

### Loading data

```python
url = "https://danskebank.com/file-cloud?filters=Investor%20relations"
```

```python
class pdfs:
    def __init__(self, url, save_fname):
        self.url = url
        self.pdf_links = None
        self.page_source = None
        self.html = None
        self.save_fname = save_fname
        
    
    def extract_links(self, wait):
        self.html = self.extract_pagesource(self.url, wait)
        
        soup = BeautifulSoup(self.html, 'html.parser')
        self.links = soup.find_all('a')
        print("Link extraction Done")

    def extract_pdflinks(self, key_phrase):
        """
        :param key_phrase: excludes all data-names that do not include key_phrase
        :type key_phrase: str
        :returns: a list of pdf-urls that includes key_phrase
        :rtype: list
        """
        pdfs = []

        for link in self.links: 
            try: 
                if ('.pdf' in link.get('href')) and (str.lower(key_phrase) in str.lower(link.get('data-name'))) and not (link.get('href') in pdfs):  
                    pdfs.append(link.get('href'))
                    #print(link.get('data-name'))
            except:
                continue
        
        self.pdf_urls = pdfs
        print("extracted pdflinks: " + str(len(pdfs)))
        
        
    def save_pdfs(self):
        did_not_finish = 0
        for pdf in tqdm(self.pdf_urls):
            try:
                #print("downloading:" + pdf)
                response = requests.get('https://danskebank.com' + pdf, verify=False)
                date = re.search('\d{4}/\d{1,2}',pdf).group(0)
                year = re.search('\d{4}', date).group(0)
                month = re.findall('/\d{1,2}', date)[-1][1:]
                open(os.path.expanduser('~') +'\\'+ self.save_fname +'\\'+ self.key_phrase.replace(" ", "_") +'_' + year +'_' + month + '.pdf', 'wb').write(response.content)

            except:
                #print("downloading:" + pdf)
                #print("did not finish")
                did_not_finish=did_not_finish + 1
                continue

        print("Incomplete downloads: " + str(did_not_finish))

    
    def download_and_save_pdfs(self,key_phrase,wait):
        
        self.key_phrase = key_phrase
        
        print("Opening browser and navigating...")
        self.extract_links(wait)
        print("Browser closed and html extracted")
        
        print("Extracting PDF-links that include: " + key_phrase)
        self.extract_pdflinks(key_phrase)
        print("Extraction Done")
        
        print("Downloading files..")
        self.save_pdfs()
        
        print("Download is done!")
        
               
                
    @staticmethod
    def extract_pagesource(url,wait):
        browser = webdriver.Chrome()
        browser.get(url)
        cookie_button = browser.find_element_by_id('button-accept-necessary')
        assert cookie_button.text == "OK to necessary"
        time.sleep(wait)
        cookie_button.click()

        cont = True
        while cont == True: 
            try:
                test = browser.find_elements_by_xpath("//body[@id='body']//main//section//div[@class='row']//div[@class='columns small-12 medium-12 large-10 large-centered']/a[@class='show-next-10 button cta']")
                test[0].click()
            except:
                cont = False

        
        page_source = browser.page_source
        browser.close()
        
        
        return page_source 
    
    
    

    
```

```python

# Initializing and choosing name of destination folder
danskebank = pdfs(url, "Reports")
# Dowloadling all reports containing the given key word(s)
danskebank.download_and_save_pdfs("Interim Report", 5)
danskebank.download_and_save_pdfs("Annual Report" , 5)
```
